```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df_agg = df.groupby('tipo_de_acidente')['N_de_feridos'].sum().reset_index()
df_agg['tipo_de_acidente'] = df_agg['tipo_de_acidente'].str.upper()
df_agg['tipo_de_acidente'] = df_agg['tipo_de_acidente'].replace(r'ATROPELAMENTO.*', 'ATROPELAMENTO', regex=True)
df_agg = df_agg.groupby('tipo_de_acidente', as_index=False).sum()
df_agg = df_agg[df_agg['N_de_feridos'] > 100]
```

```python
# Criar o gráfico de barras horizontais
plt.figure(figsize=(15, 20))
sns.barplot(data=df_agg, y='tipo_de_acidente', x='N_de_feridos', palette='viridis')

# Títulos e rótulos
plt.title('Número Total de Feridos por Tipo de Acidente', fontsize=16)
plt.xlabel('Total de Feridos', fontsize=12)
plt.ylabel('Tipo de Acidente', fontsize=12)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
def detectar_outliers_desvio_padrao(df, coluna):
    media = df[coluna].mean()
    desvio_padrao = df[coluna].std()
    outliers = df[(df[coluna] < media - 3 * desvio_padrao) | (df[coluna] > media + 3 * desvio_padrao)]
    return outliers
coluna1 = 'ilesos'
# Aplicar a função nas colunas, agora considerando 10 desvios padrão
outliers_coluna1 = detectar_outliers_desvio_padrao(df, coluna1)

print("Outliers na coluna 1:")
print(outliers_coluna1)
```
